## 1.0 Introduction and Literature Review

Fiscal decentralisation policies have been broadly suggested by the world bank to improve regional economic development (World Bank, 2008). The fiscal decentralisation policies intend to empower regional governments by giving them authority in customizing their own revenue and expenditure (Booth, 2003). It not only encourage local governments to explore and exploit their potential (Firman, 2009) but also increase their responsiveness and awareness of public need (Faguet, 2004). 

Regional economic development leads to two possible consequences: spatial equilibrium (regional convergence) and spatial disequilibrium (regional divergence or disparities) (Aritenang and Chandramidi, 2022). Local development can make regions grow apart which results in a huge economic gap that could lead to disparities. On the contrary, it gives chance for regions to utilise their potential and increase the free movement of production factors—labour and capital—which results in equal conditions with its surrounding (Lipshitz, 1992). Thus, spatial factors play an important role in regional economic development. Geographic spill-over affects  how fast regional convergence development as it has influence local neighbours (Arbia, 2006; Day and Lewis, 2013). Regional convergence is accelerated by geographic spill-over as it happens when economic development is influenced by local neighbour (Day and Lewis, 2013)

Indonesia has been aiming to reduce socioeconomic disparities and achieve regional convergence by implementing fiscal decentralisation since 2001 (Booth, 2003; Aritenang and Chandramidi, 2022). Previous studies found that fiscal decentralisation in Indonesia has positive impact on cities’ economic performance and affects regional clustering (Setiawan and Aritenang, 2019), meaning that fiscal decentralisation helps reduce regional disparities. It also confirms a previous study that stated spill-over positively affects neighbouring economic growth in Indonesia (Day and Lewis, 2013). On the other hand, using the spatial lag model, Vidyattama (2013) found a very weak regional convergence trend in Indonesia. This was confirmed by Aritenang and Chandramidi (2022) who found that Indonesia is still suffer from economic disparities in province level. The study showed that East Java, North Sumatra, Riau Island, and Papua have high gini coefficient which indicates high inequality.


## 2.0 Research Question

Although fiscal decentralisation has been implemented for over 20 years in Indonesia, it still remains unclear whether regional autonomy successfully brings Indonesia to regional equality or disparities. Thus, this research aims to assess the spatial pattern of fiscal decentralisation on a regional level in East Java which has a high gini coefficient. To achieve the objectives, this paper breakdowns into these following questions:

> **Q1** How is the distribution of local government fiscal decentralisation in East Java?
>
> **Q2** How is spatial spatial pattern of fiscal decentralisation in East Java? 
>
> **Q3** Does spill over affect the distribution of fiscal decentralisation in East Java?

## 3.0 Mothods

Previous studies used clustering to analyse the distribution of fiscal decentralisation with multiple dimensions. Clustering analysis was done to group numbers of entities into smaller groups with similar characteristics based on multiple dimensions. Considering multi-variable of fiscal decentralisation, this technique has the ability to reduce the complexity of the dataset and accurately predict the main attribute of the dataset (Liu, 2011). Thus, for this paper I used clustering methods to analyse fiscal decentralisation in East Java.

Moreover, understanding spatial patterns and spill-over effects of fiscal decentralisation can be done by exploring geographic distribution of non-spatial clusters. For this method, clustering of non-spatial variables was done first, then the assigned cluster was plotted into the maps. This method was selected to observe the spill-over effect, which cannot be seen using spatial clustering as it considers geographical adjacency in the calculation. Indication of spill-over effect was shown by agglomeration of regions with the same cluster.

K-means and DBSCAN are two non-hierarchical clustering methods that were used in this paper. Each of these methods has its own pros and cons. K-means is more popular due to its fast processing and ability to handle large datasets. Meanwhile, DBSCAN is more sensitive to outliers (Mohammed, Jacksi and Zeebaree, 2020).

### 4.1 Government Revenue Expenditure Data

In fiscal decentralisation, local governments are given the authority to manage their revenue and spending budget. Thus, assessing the impact of fiscal decentralisation can be done by examining city governments' revenue stream, which are from their own source of revenue (OSR)—tax revenue, retribution, and owned-enterprise revenue—and intergovernmental transfer (IGT) (Booth, 2003). OSR comes from tax revenue, retribution, and owned-enterprise revenue. Meanwhile, IGT is received from general or special allocation fund from higher level government—national and provincial government (Direktorat Jenderal Perimbangan Keuangan, 2017; Aritenang and Chandramidi, 2022). Previous study also indicates that government capital expenditure also the significant variables of fiscal decentralisation (Liu, 2011; Aritenang and Chandramidi, 2022)

Each city government dataset was stored in a different csv file with similar column format. However, to run analysis it is required to have one single dataset for all the cities. Thus beforehand, to read the raw file I made a list of 38 regions in East Java.

Since multiple file has to be stacked into single dataset, we will use looping `for` to read all files. The original format of the raw data is in long dataset so it has to be transpose into wide dataset.

```python
data = pd.DataFrame()

for i in city_name:
    df = pd.read_csv(
        "https://raw.githubusercontent.com/hilmanpr21/CASA0006_DSSS_fiscal-decentralization/main/data/apbd_2021/"
        + i
        + ".csv"
    )
    df = df.drop(columns=["realisasi", "persentase"])
    df.rename(columns={"akun": "index"}, inplace=True)
    df.rename(columns={"anggaran": i}, inplace=True)
    df = df.T
    if len(data) != 0:
        df = df.drop(["index"])
    data = pd.concat([data, df])

data
```

```python
# Overwrite `0` as null
data2 = data2.replace(0, np.NaN)
data2.dtypes
```

### 5.2 DBSCAN

#### Standarised The Data

From the bar plot above, some variable ranges exceed others. Although all the variables are already in the same unit, standardisation is still needed because it makes clustering less dominated by the greatest range.

```python
# extract necessary data first
raw = data2_selected2.copy()
raw.head()
```

`MinMaxScaler()` function is favourable to standardise this data that highly skewed. The data will be adjusted within range 0 to 1.

```python
mms = MinMaxScaler()

rre = raw.copy()
for c in raw.columns.values:
    rre[c] = mms.fit_transform(raw[c].values.reshape(-1, 1))
    print("The range of {} is [{}, {}]".format(c, rre[c].min(), rre[c].max()))
rre.head()
```

```python
merged.shape
```

#### Make Function for Cluster Visualisation 

As this paper aims to see spatial pattern, the cluster label is visualised after running the clustering. `mapping_cluster` is a function for plotting the cluster label into a map.

```python
def mapping_clusters(labels_cluster):
    merged["cluster_nm"] = labels_cluster
    merged.plot(
        column="cluster_nm",
        categorical=True,
        legend=True,
        figsize=(12, 8),
        cmap="Paired",
    )
    plt.title("DBSCAN Cluster Distribution")
```

#### Check Epsilom

DBSCAN has ability to find clusters in dataset that contains noise and outliers such this dataset (Khan et al., 2014). Before running DBSCAN clustering, I set a minimal number of points to exhibit cluster (minPts) and epsilon radius of cluster density (Mistry et al., 2021). Epsilon can be identified using elbow.

```python
neighbors = NearestNeighbors(n_neighbors=20)
neighbors_fit = neighbors.fit(rre)
distances, indices = neighbors_fit.kneighbors(rre)

distances = np.sort(distances, axis=0)
distances = distances[:, 1]
plt.title("Epsilon Elbow Chart")
plt.plot(distances)
```

The elbow showed 0.15, which means that it is the suggested epsilon radius for nearby point to cluster.

Min points should be larger than the number of data dimensions. Previous study suggests using min points that equals to twice the number of variables (Sander et al., 1998). However, the dataset has a small observation number, so I decided to use `minPts = dimension+1` which was 6.

```python
# run DBSCAN
minPts = 6  
epsilon = 0.15
dbsc = DBSCAN(eps=epsilon, min_samples=minPts)
dbsc.fit(rre)

# extract the cluster label into new object to be visualized
cluster_nm = dbsc.labels_

# visualise the cluster into maps
mapping_clusters(cluster_nm)

# Save cluster label and original dataset into 1 dataset
# rre_dbscan = rre.copy()
# rre_dbscan["dbscan_labels"] = dbsc.labels_

# show number of entries per cluster
pd.Series(dbsc.labels_).value_counts()
```

The DBSCAN clustering showed 2 clusters, assigned as 0 and 1, with members of 22 and 7 respectively. Meanwhile, 9 regions were identified as noise which means it does not belong to any cluster. It happened because these regions are considered outliers due to their huge distance.

Cluster 0 shows agglomeration in Madura island (upper north), east south, and west area. However, every cluster 1 is surrounded by cluster 0 which indicates that geographic spill over does not occur.

### 5.4 K-means

Since DBSCAN identified a lot of regions as noise, the next step was to try to use K-means. As this method is less robust to outliers, Surabaya city which is identified as an outlier was taken out from the dataset.

#### Drop Outliers

```python
# drop `surabaya` from fiscal dataset
rre2 = rre.drop(["Surabaya"])
rre2
```

```python
# drop `surabaya` which is in index `34` from geometry dataset
merged2 = merged.drop([34])
merged2
```

#### Make Visualisation Function

```python
def mapping_clusters2(labels_cluster):
    merged2["cluster_nm"] = labels_cluster
    merged2.plot(
        column="cluster_nm",
        categorical=True,
        legend=True,
        figsize=(12, 8),
        cmap="Paired",
    );
```

#### Find Number of Clustering

K-means clustering requires identifying the number of clusters (k-value) beforehand. Typically, k-value is identified from the elbow plot with SSE (Sum of Squared Errors). SSE is a performance indicator of clustering. Smaller SSE indicates better performance. The elbow shows a balance value between small SSE and small k-value.

```python
# calculate SSE for a range of number of cluster
list_SSE = []
min_k = 1
max_k = 10
range_k = range(min_k, max_k + 1)
for i in range_k:
    km = KMeans(
        n_clusters=i, init="random", n_init=10, max_iter=300, tol=1e-04, random_state=0
    )
    km.fit(rre2)
    # inertia is a concept from physics. Roughly it means SSE of clustering.
    list_SSE.append(km.inertia_)

# plot
plt.plot(range_k, list_SSE, marker="o")
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
plt.show()
```

The elbow plot suggests 3 clusters.

```python
k_cluster = 3
random_seed = 100
kmeans_method = KMeans(n_clusters=k_cluster, random_state=random_seed)
kmeans_method.fit(rre2)

# plotting
mapping_clusters2(kmeans_method.labels_)

#  Save cluster label and original dataset into 1 dataset
# rre_kmeans = rre2.copy()
# rre_kmeans["kmeans_labels"] = kmeans_method.labels_

pd.Series(kmeans_method.labels_).value_counts()
```

K-means clustering result shows that cluster 1 has the biggest number with 23 regions, meanwhile cluster 0 and 2 have 7 members. Although members of cluster 1 are adjacent to each other, we cannot say it indicates spill-over effects because members of cluster 0 and 2 are geographically spread.

#### Compare Performance of K-means and DBSCAN

```python
# Count the SSE value of DBSCAN
metrics.silhouette_score(rre, dbsc.labels_)
```

```python
# Count the SSE value of K-means
metrics.silhouette_score(rre2, kmeans_method.labels_)
```

|Clustering Methods|SSE |

DBSCAN has lower SSE value which means that it has better performance to cluster this dataset. Thus, for future research on fiscal decentralisation, it is suggested to use DBSCAN clustering.
